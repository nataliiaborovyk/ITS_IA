```python
# Esercizio 1
# Scrivi il codice di Aggiornamento Bayesiano sequenziale
# Stima della probabilità di una email di Spam e se supera la soglia predefenita (ad es 0.7)
# scrivi se è o non è una mail di Spam
# Prior: P(spam)= 0.4
# Osserviamo una delle parole in una email ed aggiorniamo la probabilità

# Likelihood P(parola | Spam) e P(parola | Non-Spam)
lik_spam = {
    "win":   0.7,
    "meeting":0.05,
    "offer": 0.6,
    "hello": 0.02,
    "prize": 0.8,
    "project":0.04,
    "update": 0.10,
    "invoice": 0.01
}

lik_nonspam = {
    "win":   0.1,
    "meeting":0.2,
    "offer": 0.05,
    "hello": 0.3,
    "prize": 0.08,
    "project":0.25,
    "update": 0.15,
    "invoice": 0.4
}

# sequenza parole osservate
parole = ["win", "meeting", "project", "offer", "prize", "win"]

# Prior iniziale
p_spam = 0.4

# Esercizio 2
# Applicando il codice dell'esercizio precedente, modifica la sequenza delle parole
# e verifica come varia la probabilità di spam


# Soglia per decidere se è spam
soglia_spam = 0.7

def aggiorna_prob_spam(parole, prior_spam):
    """
    parole: lista di parole osservate nella email
    prior_spam: probabilità iniziale P(spam)
    """
    p_spam = prior_spam  # iniziamo dalla prior

    print(f"Probabilità iniziale P(spam) = {p_spam:.4f}")
    print("-" * 50)

    # Cicliamo su ogni parola osservata nella email
    for parola in parole:
        # Prendiamo le likelihood P(parola | spam) e P(parola | non-spam)
        p_word_given_spam = lik_spam[parola]
        p_word_given_nonspam = lik_nonspam[parola]

        # Calcoliamo P(non-spam) come 1 - P(spam)
        p_nonspam = 1 - p_spam

        # Calcoliamo il denominatore (evidenza) P(parola)
        evidenza = p_word_given_spam * p_spam + p_word_given_nonspam * p_nonspam

        # Applichiamo Bayes: P(spam | parola)
        p_spam = (p_word_given_spam * p_spam) / evidenza

        # Stampiamo lo stato dopo aver visto questa parola
        print(f"Parola osservata: {parola}")
        print(f"  P(parola | spam)      = {p_word_given_spam:.4f}")
        print(f"  P(parola | non-spam)  = {p_word_given_nonspam:.4f}")
        print(f"  Nuova P(spam | dati)  = {p_spam:.4f}")
        print("-" * 50)

    # Alla fine ritorniamo la probabilità finale di spam
    return p_spam

# ======================
# ESERCIZIO 1
# ======================

# Sequenza parole osservate (data dal testo dell'esercizio 1)
parole_es1 = ["win", "meeting", "project", "offer", "prize", "win"]

# Prior iniziale
prior_iniziale = 0.4

# Calcoliamo la probabilità finale di spam dopo aver visto tutte le parole
p_finale_spam_es1 = aggiorna_prob_spam(parole_es1, prior_iniziale)

print(f"Probabilità finale di SPAM (esercizio 1) = {p_finale_spam_es1:.4f}")

# Classificazione finale
if p_finale_spam_es1 > soglia_spam:
    print("Risultato: la email è classificata come SPAM")
else:
    print("Risultato: la email NON è classificata come spam")

# ======================
# ESERCIZIO 2
# ======================

# Adesso cambiamo la sequenza delle parole (come chiede l'esercizio 2)
# Puoi provare varie sequenze, qui ne metto una di esempio
parole_es2 = ["hello", "meeting", "project", "update", "invoice"]

p_finale_spam_es2 = aggiorna_prob_spam(parole_es2, prior_iniziale)

print(f"Probabilità finale di SPAM (esercizio 2) = {p_finale_spam_es2:.4f}")

if p_finale_spam_es2 > soglia_spam:
    print("Risultato: la email è classificata come SPAM")
else:
    print("Risultato: la email NON è classificata come spam")



# Esercizio 3
# Scrivi il codice di Aggiornamento Bayesiano sequenziale
# che stima della probabilità che un paziente sia malato o sano in funzione
# del risultato di una serie di esami che danno valore positivo o negativo
# ---------------------------------------------------------
# Scenario clinico: un paziente esegue una serie di test
# ---------------------------------------------------------

# Prevalenza (prior): P(Malattia)
p_malattia = 0.02   # 2%

# Test diagnostico (uguale per tutti i test)
sensibilita = 0.92               # P(Test+ | Malato)
specificita = 0.95               # P(Test- | Sano)
falso_positivo = 1 - specificita # P(Test+ | Sano)
falso_negativo = 1 - sensibilita # P(Test- | Malato)

# Sequenza reale dei test effettuati
test = ['+', '-', '+', '+', '-']

# Storia della probabilità da plottare
storia_probabilita = [p_malattia]
```
